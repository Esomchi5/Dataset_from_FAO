```python
# question 13
data_new = data.drop(columns=["date", "lights"])
#data_new.head()

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

normalised_data = pd.DataFrame(scaler.fit_transform(data_new), columns=data_new.columns)
features = data_new.drop(columns=['Appliances'])
target = data_new['Appliances']


from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
linear_model1 = LinearRegression()

linear_model1.fit(x_train, y_train)

predicted_values = linear_model1.predict(x_test)
```

```python
# questions 13, 14, 15, 16

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, predicted_values)
rmse= np.sqrt(mean_squared_error(y_test, predicted_values))
rss = np.sum(np.square(y_test - predicted_values))
r2e = r2_score(y_test, predicted_values)

print(round(mae, 2))
print(round(rss, 2))
print(round(rmse, 3))
print(round(r2e, 2))
```

```python
 # question 17
    
def get_weights_df(model, fit, col_name):
#this function returns the weight of every feature
    weights = pd.Series(model.coef_, fit.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df

Weights = get_weights_df(linear_model1, x_train, 'Linear_Model_Weight')

Weights
```

```python
#qestion 18

from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha=.4)
ridge_reg.fit(x_train, y_train)

predicted_v = ridge_reg.predict(x_test)
from sklearn.metrics import mean_absolute_error

mae1 = mean_absolute_error(y_test, predicted_v)
print(round(mae1, 2))
```

```python
# question 19
from sklearn.linear_model import Lasso

lasso_reg = Lasso(alpha=.001)
lasso_reg.fit(x_train, y_train)

def get_weights_df(model, fit, col_name):
#this function returns the weight of every feature
    weights = pd.Series(model.coef_, fit.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df

Weights = get_weights_df(lasso_reg, x_train, 'Linear_Model_Weight')

Weights
```

```python
# question 20

from sklearn.metrics import mean_squared_error

predicted = lasso_reg.predict(x_test)
rmse= np.sqrt(mean_squared_error(y_test, predicted))

print(round(rmse, 3))
```
